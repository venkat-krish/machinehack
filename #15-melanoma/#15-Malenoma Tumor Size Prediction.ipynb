```python
train_data = pd.read_csv("./data/Train.csv")
test_data = pd.read_csv("./data/Test.csv")
submit_data = pd.read_csv("./data/sample_submission.csv")
```

```python
def view_data(dataframe):
    dataframe.info()
    print("--"*30)
    print(dataframe.head().append(dataframe.tail()))
    
# Calculating Multicollinearity score
def compute_mulcol_score(dataframe, score_range):
    mul_score = pd.DataFrame()
    mul_score['Variables'] = dataframe.columns
    mul_score['VIF'] = [variance_inflation_factor(dataframe.values, i) 
                         for i in range(dataframe.shape[1])]
    
    # Score that are less than 5 will be selected
    non_mul_col = [item[0] for item in mul_score.values if item[1] < score_range]

    return non_mul_col
    
```

## Building a model

```python
# Predict the test data
test_data['clusterid'] = kmeans.predict(X_test_scaled)
```

```python
# feature_cols = ['mass_npea', 'size_npear', 'malign_ratio', 'damage_size', 'exposed_area', 'std_dev_malign', 'err_malign', 'malign_penalty', 'damage_ratio']
feature_cols = ['size_npear', 'malign_ratio', 'std_dev_malign','damage_size', 'err_malign', 'malign_penalty', 'damage_ratio']

data_0 = train_data[train_data['clusterid'] == 0]
data_1 = train_data[train_data['clusterid'] == 1]

X_train_0 = data_0[feature_cols]
X_train_1 = data_1[feature_cols]
y_train_0 = data_0['tumor_size']
y_train_1 = data_1['tumor_size']
```

```python
# Test data
test_data_0 = test_data[test_data['clusterid'] == 0]
test_data_1 = test_data[test_data['clusterid'] == 1]

X_test_0 = test_data_0[feature_cols]
X_test_1 = test_data_1[feature_cols]
```

```python
scaler_0 = StandardScaler().fit(X_train_0)

X_train_0[feature_cols] = scaler_0.transform(X_train_0)
X_test_0[feature_cols] = scaler_0.transform(X_test_0)

scaler_1 = StandardScaler().fit(X_train_1)

X_train_1[feature_cols] = scaler_0.transform(X_train_1)
X_test_1[feature_cols] = scaler_0.transform(X_test_1)
```

```python
X_test_0.columns
```

```python
# Merge the data
xgb_0 = XGBRegressor(colsample_bytree=0.4,
                 gamma=0,                 
                 learning_rate=0.1,
                 max_depth=5,
                 min_child_weight=1.5,
                 n_estimators=10000,                                                                    
                 reg_alpha=0.75,
                 reg_lambda=0.45,
                 subsample=0.6,
                 seed=42)
xgb_0.fit(X_train_0,y_train_0)

xgb_1 = XGBRegressor(colsample_bytree=0.4,
                 gamma=0,                 
                 learning_rate=0.1,
                 max_depth=5,
                 min_child_weight=1.5,
                 n_estimators=10000,                                                                    
                 reg_alpha=0.75,
                 reg_lambda=0.45,
                 subsample=0.6,
                 seed=42)
xgb_1.fit(X_train_1, y_train_1)
```

```python
print(xgb_0.score(X_train_0, y_train_0))
print(xgb_1.score(X_train_1, y_train_1))
```

## Predict the test data

```python
X_test_0['pred_size'] = xgb_0.predict(X_test_0)
X_test_1['pred_size'] = xgb_1.predict(X_test_1)
```

```python
final_preds = X_test_0.append(X_test_1)
final_preds = final_preds.sort_index(axis = 0)
```
